#### Data preparation

```python
# Importing pandas library
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Load the dataset into a pandas DataFrame
df = pd.read_csv(r"C:\Users\acer\Desktop\New folder\Training Data Dump - Dump.csv")
df.head()
```

## Data cleaning (wrangling)

###### checking no of rows and columns 

```python
df.shape
```

###### checking null values

```python
df.isnull()
```

```python
# Checking out distribution of null values
df.isnull().sum()
df.head(10)
```

###### As we can see there are so many null values in F2,F4, F5 and so on we need to remove it..

```python
data=df.dropna(axis=1) # droping null values column wise
data.head()
```

```python
data.head()
```

```python
# Removing column F3 in from dataset as it's not required as per prediction
data.drop('F3',axis=1,inplace=True)
data.head()
```

```python
x = data.drop(columns=['Label'])  # Features
y = data['Label'] # Target col
```

## Data splitting

###### Splitting our datasets into training and testing part 
###### test_size=0.2 tells us that our testing data is 20% of whole data, remaining 80% is our training data
###### We can assign test_size =0.3 as per our suitability.
###### random_state: We are getting diffrent slots after each execution hence to avoid this shuffling
###### random_state play important role , it's mandatory to assign integer value like 5, 42 etc.
######  Otherwise it would n't work.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

######  Here max_iter =1000 means our algorithms can execute upto 1000 time that might help to come up with optimal solution
###### Sometime huge data needs more iterations to converge solutions.

## Model selection and evaluation

```python
# Fitting our training data in our logestic regression model.

logistic_regression = LogisticRegression(max_iter=1000, random_state=42)
logistic_regression.fit(x_train, y_train)
```

#### Predictions

```python
y_pred_lr = logistic_regression.predict(x_test)
```

#### Metrices for logistic regression

```python
# Accuracy
Accuracy_score=accuracy_score(y_test, y_pred_lr)
# Precision
Precision_score= precision_score(y_test, y_pred_lr)
# Recall
Recall_score=recall_score(y_test, y_pred_lr))
# f1_score
F1_score=f1_score(y_test, y_pred_lr))
```

#### DecisionTreeClassifier

```python
dtree_clf = DecisionTreeClassifier(random_state=42)
# Fitting training data into decision tree classifier model
dtree_clf.fit(x_train, y_train)
```

####  Predictions

```python
y_predict_decision_tree=dtree_clf.predict(x_test)
```

#### Metrices for DecisionTreeClassifier 

```python
# Accuracy
Accuracy_score=accuracy_score(y_test, y_predict_decision_tree)
# Precision
Precision_score= precision_score(y_test, y_predict_decision_tree)
# Recall
Recall_score=recall_score(y_test, y_predict_decision_tree))
# f1_score
F1_score=f1_score(y_test, y_predict_decision_tree))
```

## Model explainability

#### Logistic regression
###### logistic regression is best for linear relationships between features and the target variable like LinearRegression.
###### It works well with smaller datasets and is less prone to overfitting.
###### Overfitting means your predicting values covers your all data points which is an ideal case.

#### DecisionTree model
###### Decisiontree models handles non-linear relationships well and can lead to complex patterns and hence more prone
###### To overfitting.

#### NOTE:
###### Depending on size of data we choose our model like for huge data we might prefer decision tree whereas for medium size
###### Data we can go for logistic regression model
###### We neither want overfitting(giving large parameters to your model) nor want underfitting(giving less parameters to your model)
###### In both case model might predict wrong outputs, we want best_fit curve(giving suitable amount of parameters) so
###### That our model can give more accurate values or results.
